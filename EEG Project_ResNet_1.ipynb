```python
X_fin = pd.DataFrame(data=X_trans)
X_fin.head(10)
```

```python
X_fin=X_fin.values
y=y.values
```

```python
import numpy as np
from keras.layers import Input, Conv2D, BatchNormalization, Activation, MaxPooling2D, Flatten, Dense, Dropout
from keras.models import Model
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split

# Define the Residual Unit
def residual_unit(x, filters, strides=1, kernel_size=3, activation='relu'):
    y = Conv1D(filters, kernel_size=kernel_size, strides=strides, padding='same')(x)
    y = BatchNormalization()(y)
    y = Activation(activation)(y)
    y = Conv1D(filters, kernel_size=kernel_size, padding='same')(y)
    y = BatchNormalization()(y)
    if strides !=1 or x.shape[-1] != filters:
        x = Conv1D(filters, kernel_size=1, strides=strides, padding='same')(x)
    return Activation(activation)(x + y)

# Build the ResNet Model
def build_resnet(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    x = Conv1D(64, kernel_size=7,strides=2, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=3, strides=2, padding='same')(x)

    x = residual_unit(x, filters=64)
    x = residual_unit(x, filters=64)

    x = residual_unit(x, filters=128, strides=2)
    x = residual_unit(x, filters=128)

    x = residual_unit(x, filters=256, strides=2)
    x = residual_unit(x, filters=256)

    x = residual_unit(x, filters=512, strides=2)
    x = residual_unit(x, filters=512)

    x = MaxPooling1D(pool_size=2)(x)
    x = Flatten()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

# Load your EEG dataset (replace this with your actual data loading code)
eeg_data =X_fin
labels =y

# Assuming EEG data is already preprocessed and formatted appropriately
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(eeg_data, labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)

# Reshape EEG data to fit the ResNet input shape
input_shape = (eeg_data.shape[1],1)

# Build the ResNet model
num_classes = 3
model = build_resnet(input_shape, num_classes)

#metrics
METRICS = [tf.keras.metrics.CategoricalAccuracy(name='acc'),
           tf.keras.metrics.AUC(name='auc'), 
           tfa.metrics.F1Score(num_classes=3)]
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=METRICS)

# Define callbacks
#checkpoint = ModelCheckpoint('resnet_model.h5', monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)


# Train the model
history=model.fit(X_train, y_train, batch_size=32, epochs=50, validation_data=(X_test, y_test))
```
